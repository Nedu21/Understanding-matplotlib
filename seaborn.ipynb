```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme(style='whitegrid')
sns.set_palette("Accent")
```

```python
# Create example DataFrame
df = pd.DataFrame({
    'x': np.linspace(0, 10, 100),
    'y': np.sin(np.linspace(0, 10, 100)) + np.random.normal(0, 0.1, 100)
})

# Seaborn scatter plot
sns.scatterplot(x='x', y='y', data=df)
plt.title("Seaborn Scatter Plot")
plt.show()
```

```python
# Seaborn line plot with smoothing (optional)
sns.lineplot(x='x', y='y', data=df, color='coral')
plt.title("Seaborn Line Plot")
plt.show()
```

```python
sns.regplot(x='x', y='y', data=df)
plt.title("Scatter with Regression Line (Seaborn)")
plt.show()
```

```python
tips = pd.read_csv('./data/tips_simulated.csv')

tips.head()
```

```python
sns.countplot(x='day', data=tips, hue='day', palette='pastel')
plt.title("Count of Restaurant Bills by Day")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, errorbar=None, hue='sex', palette='muted')
plt.title("Average Bill per Day")
plt.show()
```

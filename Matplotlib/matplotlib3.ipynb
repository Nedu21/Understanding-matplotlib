```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Boxplots, Violin Plots, and Multiple Subplots

```python
scores_class_A = np.random.normal(70, 10, 50)
scores_class_B = np.random.normal(75, 12, 50)

plt.figure(figsize=(10, 4))
plt.boxplot([scores_class_A, scores_class_B], patch_artist=True)
plt.xticks([1, 2], ['Class A', 'Class B'])
plt.title('Exam scores by class')
plt.ylabel('Score')
plt.show()
```

```python
import seaborn as sns

# Example data: again two classes
data = {
    'score': np.concatenate([scores_class_A, scores_class_B]),
    'class': ['A']*len(scores_class_A) + ['B']*len(scores_class_B)
}
sns.violinplot(x='class', y='score', data=data, palette='Pastel1')
plt.title("Violin Plot of Exam Scores")
plt.show()
```

```python
# Create a figure with 1 row, 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Left plot: Boxplot
axes[0].boxplot([scores_class_A, scores_class_B], tick_labels=['A', 'B'])
axes[0].set_title("Boxplot of Scores")

# Right plot: Histogram
axes[1].hist(scores_class_A, bins=8, color='lightgray', edgecolor='black', alpha=0.7)
axes[1].set_title("Histogram of Class A Scores")
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

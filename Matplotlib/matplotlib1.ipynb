```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

# Line charts

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(10, 4))
plt.plot(x, y, marker = '+', linestyle = '-')
plt.title('Simple sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

# Scatter plots

```python
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

plt.figure(figsize=(10,4))
plt.scatter(x, y, marker='o', color='cyan', edgecolor='black')
plt.title('Scatter plot example')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.grid(True)
plt.show()
```

# Exercises

## 1 Plot a lineplot

```python
df = pd.read_csv('./data/flights.csv')
df.columns = ['Year', 'Month', 'Passengers']
df.head()
```

```python
yr_sum = df.groupby('Year').sum()
yr_sum
```

```python
plt.figure(figsize=(10, 4))
plt.plot(yr_sum.index, yr_sum['Passengers'], marker='+', linestyle='-')
plt.title('A plot of total passengers each year')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.grid(True)
plt.show()
```

## 2 Plot a scatterplot

```python
np.random.seed(42)
study_hours = np.random.uniform(0, 15, 100)
exam_scores = 6 * study_hours + np.random.normal(0, 8, 100)
# Optional: cap the scores at 100
exam_scores = np.clip(exam_scores, 0, 100)

plt.figure(figsize=(10, 4))
plt.scatter(study_hours, exam_scores, marker='o', color='blue', edgecolors='black')
plt.title('Study hours vs exam score')
plt.xlabel('Study hours')
plt.ylabel('Exam score')
plt.grid(True)
plt.show()
```

## 3 Plot a Real Time Series

```python
dates = [datetime.datetime(2023, 1, i+1) for i in range(10)]
values = np.random.randint(10, 20, size=10)

plt.figure(figsize=(10, 4))
plt.plot(dates, values, marker='o', linestyle='-.')
plt.gcf().autofmt_xdate()  # Format the date labels nicely
plt.title('Time plot series')
plt.xlabel('Dates')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Histograms, Bar Charts, and Pie Charts

```python
np.random.seed(0)
ages = np.random.normal(loc=30, scale=5, size=100).astype(int)

plt.figure(figsize=(10, 4))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of ages')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
clubs = ['Real madrid', 'AC milan', 'Bayern', 'Liverpool', 'Barcelona', 'Ajax', 'Manchester united', 'Inter milan', 'Chelsea']
ucl_titles = [15, 7, 6, 6, 5, 4, 3, 3, 2]

plt.figure(figsize=(10, 4))
plt.title('UCL trophies for each team')
plt.bar(clubs, ucl_titles, color='salmon')
plt.xticks(rotation=20)
plt.xlabel('Clubs')
plt.ylabel('UCL trophies')
plt.grid(True)
plt.show()
```

```python
categories = ['Region 1', 'Region 2', 'Region 3']
sales_A = [20, 35, 30]
sales_B = [25, 30, 45]
x = np.arange(len(categories))

plt.figure(figsize=(10,4))
width = 0.35
plt.bar(x - width/2, sales_A, width, label='Product A')
plt.bar(x + width/2, sales_B, width, label='Product B')
plt.xticks(x, categories)
plt.ylabel("Sales")
plt.title("Product Sales by Region")
plt.legend()
plt.show()
```

```python
budget = ['Savings', 'Investments', 'Birthday', 'Transportation', 'Rent', 'Feeding', 'Internet', 'Miscellaneous']
amounts = [600000, 450000, 150000, 354000, 708000, 531000, 30000, 177000]

plt.figure(figsize=(10, 4))
plt.pie(amounts, labels=budget, autopct='%1.1f%%', shadow=True)
plt.title("June budget")
plt.show()
```

# Exercises

```python
titanic = pd.read_csv('../data/titanic.csv')
iris = pd.read_csv('../data/iris.csv')
```

```python
titanic.replace('', np.nan, inplace=True)
titanic.dropna(subset=['age', 'embark_town'], axis=0, inplace=True)
titanic.reset_index()
titanic.head(15)
```

```python
iris.head()
```

## 1 Plot a histogram


```python
plt.figure(figsize=(10, 4))
plt.hist(titanic['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of ages')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## 2 Plot a bar chart

```python
counts = iris['species'].value_counts()
species = counts.index
freq = counts.values

plt.figure(figsize=(10, 4))
plt.bar(species, freq, color='salmon')
plt.title('Bar chart')
plt.xlabel('Species')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
count = titanic['embark_town'].value_counts()
towns = count.index
frequency = count.values

plt.figure(figsize=(10, 4))
plt.bar(towns, frequency, color='thistle', edgecolor='black')
plt.title('Bar chart of towns')
plt.xlabel('Towns')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## 3 Plot a pie chart

```python
survivor = titanic['alive'].value_counts()
survivor_count = survivor.values
survived = survivor.index
explode = [0, 0.05]

plt.figure(figsize=(10, 4))
plt.pie(survivor_count, labels=survived, autopct='%1.1f%%', explode=explode, shadow=True)
plt.title('Survival rate')
plt.show()
```
